```python
from scholarly import scholarly
from scholarly import ProxyGenerator

import pandas as pd
import time
import random

```

```python


def setup_proxy():
    pg = ProxyGenerator()
    success = pg.FreeProxies()
    if success:
        scholarly.use_proxy(pg)
    else:
        print("Failed to set up proxy.")
    return success

def search_scholar(keywords, start_year, end_year):
    search_query = scholarly.search_pubs(keywords)
    
    results = []
    for result in search_query:
        try:
            # Introduce a random delay to avoid being blocked
            time.sleep(random.uniform(1, 3))
            
            pub_year = result.get('bib', {}).get('pub_year', 0)
            try:
                pub_year = int(pub_year)
                if start_year <= pub_year <= end_year:
                    data = {
                        'title': result.get('bib', {}).get('title', ''),
                        'doi': result.get('bib', {}).get('doi', ''),
                        'authors': ', '.join(result.get('bib', {}).get('author', [])),
                        'year': pub_year,
                        'month': result.get('bib', {}).get('pub_month', ''),
                        'abstract': result.get('bib', {}).get('abstract', ''),
                        'citation': result.get('num_citations', 0),
                        'publication': result.get('bib', {}).get('journal', '')
                    }
                    results.append(data)
            except ValueError:
                continue  # Skip entries without valid publication year
        except StopIteration:
            break

    return results

def save_to_csv(data, filename):
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False)
    print(f"Data saved to {filename}")

if __name__ == "__main__":
    if setup_proxy():
        keywords = 'triatomine United States'#input("Enter search keywords: ")
        start_year = 2022#int(input("Enter start year: "))
        end_year = 2024#int(input("Enter end year: "))
        filename = '/Users/liting/Documents/GitHub/google-scholar-search/north_america_triatomine/google_scholar_raw_US.csv'#input("Enter output CSV filename: ")


        results = search_scholar(keywords, start_year, end_year)
        save_to_csv(results, filename)
    else:
        print("Proxy setup failed. Please try again later or use a different method.")


```
